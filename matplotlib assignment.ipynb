```python
"""
Scipy:
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
Expected Output:

"""
```

```python
import numpy as np
from scipy.optimize import curve_fit
import scipy as sp
help(sp.optimize)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.style.use(['ggplot'])
plt.subplots(figsize=(8,6))
             
months = np.arange(12)

TMax =np.array([ 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
TMin =np.array([ 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
plt.plot(months,TMax,'ro')
plt.plot(months,TMin,'bo')
plt.xlabel('Months')
plt.ylabel('Min and max temperature')

plt.show()
```

```python
import numpy as np
def yeartemp(data,a,b,c):
    return (a + b * np.cos((data + c) * 2 * np.pi / data.max()))

r_max, c_max = sp.optimize.curve_fit(yeartemp, months,TMax, [40, 20, 0])
r_min, c_min = sp.optimize.curve_fit(yeartemp, months,TMin, [-40, 20, 0])
```

```python
r_max, c_max
```

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,6))
plt.plot(months, TMax, 'ro')
plt.plot(days, yeartemp(days, *r_max), 'r-')
plt.plot(months, TMin, 'bo')
plt.plot(days, yeartemp(days, *r_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
